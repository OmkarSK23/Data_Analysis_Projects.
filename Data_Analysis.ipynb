```python
import pandas as pd
import numpy as np

df = pd.read_csv("analysis_dataset.csv")
df
```

Now I will start with the first process of Data Cleaning.


Following Process include

*   Identify the Missing values
*   Handle Missing Values
*   Remove Duplicate Records
*   Handle Inconsistent Values






```python
#Identifuy the Missing Values

df.isnull().sum()
```

```python
# Handle Missing Values

df["Age"].fillna(df["Age"].mean(), inplace=True)
# df["City"].fillna("Unknown", inplace=True)
# df['City'] = df['City'].fillna(df['City'].mode()[0])

df['Age']

```

```python
df = df.dropna(axis=1, how='all')
df
```

```python
df['City'] = df['City'].fillna(df['City'].mode()[0])
df
```

```python
# Remove Duplicate Records

df.drop_duplicates(inplace=True)
df
```

```python
# Handle Inconsistent Values

df['Gender'] = df['Gender'].str.lower().replace({
    'm': 'Male',
    'f': 'Female'
}).str.capitalize()
df
```

The Data Transformation includes the below following:
*   Convert Data Types
*   Handle Invalid & Anomalous Values



```python
# Convert the Data Types
df["CustomerID"] = df["Customer_ID"].astype(int)
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df["Purchase_Amount"] = pd.to_numeric(df["Purchase_Amount"], errors="coerce")
df
```

```python
# Handle Invalid & Anomalous Values

# Remove negative values
df = df[df["Quantity"] >= 0]
df = df[df["Purchase_Amount"] > 0]

# Cap extreme outliers
df["Quantity"] = df["Quantity"].clip(upper=50)
df
```

Feature Engineering (Transformation)

```python
df["TotalValue"] = df["Quantity"] * df["Purchase_Amount"]
```

```python
# Summary statistics
df.describe()
```

```python
# Aggregation
df.groupby("City")["TotalValue"].sum()
```

```python
df.groupby("Gender")["TotalValue"].mean()
```

```python
# Average Purchase by payment mode
df.groupby("Payment _Mode")["Purchase_Amount"].mean().plot(
    kind="bar",
    title="Average Purchase by Payment Mode"
)

```

```python
# Multiple Metrics Together
df.groupby("City")[["Purchase_Amount", "Quantity"]].mean().plot(
    kind="bar",
    title="Average Purchase & Quantity by City"
)
```
